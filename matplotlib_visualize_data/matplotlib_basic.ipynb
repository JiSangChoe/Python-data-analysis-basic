### 차트의 종류

**막대 그래프**: 데이터의 카테고리별 값을 막대로 나타내는 방식(수직 막대 그래프, 수평막대 그래프)
> 카테고리별 비교할 때 주로 사용됨(제품별 판매량, 지역별 인구수)

**선 그래프**: 시간의 흐름에 따른 데이터 포인터를 선으로 연결하여 나타내는 방식
> 시간의 흐름에 따라 비교할 때 주로 사용됨(기온의 변화, 주식 가격)

**파이 차트**: 전체에 대한 비율을 원형으로 나타내는 방식(각 조각의 크기가 해당 부분의 비율)
> 전체에 대한 비율을 비교할 때 주로 사용됨(여론조사 결과)

**산점도**: 두 변수 간의 상관 관계를 점들의 분포로 나타내는 방식
> 두 변수의 상관관계를 파악할 때 주로 사용됨(키와 몸무게의 상관관계)

**히스토그램**: 범위별로 그룹화해서 각 그룹에 속하는 데이터 포인터의 수를 막대로 나타내는 방식
> 데이터의 분포를 비교할 때 주로 사용됨(시험 점수의 분포) 

**영역 그래프**: 선 그래프의 아래 영역에 색으로 채운 형태
> 시간의 흐름에 따라 비교할 때 주로 사용됨, 변화량을 강조할 때 주로 사용됨(누적 감염자, 누적수출량)

**박스 그래프**: 데이터의 분포를 나타내는 방식, 최소값,최대값,사분위를 나타내는 방식
> 데이터의 이상치를 판단할 때 주로 사용됨()

### 차트 그림의 구조
matplotlib 으로 그려지는 차트는 Figure 객체, Axes 객체, Axis 객체로 구성이 되어져 있음
Figure 객체는 하나 이상의 Axes 객체를 포함하고 있고, Axes 객체는 두개 이상의 Axis 객체를 포함하고 있음
- `Figure`: 차트가 그려지는 캔버스
- `Axes`: 하나의 플롯(그래프)
- `Axis`: 하나의 축

#### Figure 객체

### `Axes` 객체와 subplot 
하나의 Figure 객체 안에 여러개의 Axes 객체를 생성하여 행렬 형태의 서브 플롯을 생성할 수 있음
이때, `subplot()` 메서드를 사용하여 명시적으로 Axes 객체를 생성할 수 있음
그러나, pyplot의 `plot()` 메서드를 바로 사용해도 Axes 객체를 생성할 수 있음

```python
import numpy as np
```

```python
x1 = np.linspace(0,5)
x2 = np.linspace(0,2)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2,1,1)
plt.plot(x1,y1,'yo-')
plt.title('2행 1열 서브 플롯 첫 번째')

plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
plt.title('2행 1열 서브 플롯 두 번째')

#tight_layout 명령을 실행하면 플롯간의 간격을 자동으로 맞춰준다.
plt.tight_layout()
plt.show()
```

`subplot()` 메서드의 (x, y, index) 의 인수를 (xyindex)로 붙여서 사용할 수 있음
예시 : `subplot(2,2,1)` -> `subplot(221)`

```python
plt.subplot(221)
plt.plot(np.random.rand(5))
plt.title('첫 번째 플롯')

plt.subplot(222)
plt.plot(np.random.rand(5))
plt.title('두 번째 플롯')

plt.subplot(223)
plt.plot(np.random.rand(5))
plt.title('세 번째 플롯')

plt.subplot(224)
plt.plot(np.random.rand(5))
plt.title('네 번째 플롯')

plt.tight_layout()
plt.show()
```

`subplots(x, y)` 메서드로 복수의 axes 객체를 담은 배열을 생성할 수 있음
`subplots(x, y)` 메서드는 (figure, axes 배열) 튜플을 반환

```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(np.random.rand(5))
axes[0, 0].set_title('1행 1열')

axes[0, 1].plot(np.random.rand(5))
axes[0, 1].set_title('1행 2열')

axes[1, 0].plot(np.random.rand(5))
axes[1, 0].set_title('2행 1열')

axes[1, 1].plot(np.random.rand(5))
axes[1, 1].set_title('2행 2열')


plt.tight_layout()
plt.show()
```

#### Axis 객체와 축
`twinx()` 메서드로 x 축을 공유하는 새로운 Axes 객체를 생성할 수 잇음

```python
# 여러가지 플롯을 하나의 Axes 객체에 표시할 때 y값의 크기가 달라서 표시하기 힘든 경우가 있다. 이 때는 다음처럼 twinx 명령으로 복수의 y 축을 가진 플롯을 만들수도 있다. twinx 명령은 x 축을 공유하는 새로운 Axes 객체를 만든다.

figure,axes0 = plt.subplots()
axes1 = axes0.twinx()

axes0.set_title('x 축을 공유하는 플롯')
axes0.plot([10, 5, 2, 9, 7],'r-',label='y0')
axes0.set_ylabel('y0')

axes1.plot([100, 200, 220, 180, 120],'g:',label='y1')
axes1.set_ylabel('y1')

axes0.set_xlabel('x 축')
plt.show()
```
